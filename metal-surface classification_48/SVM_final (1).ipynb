```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import cv2
```

```python
from keras.models import Model, Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.layers import BatchNormalization
import os
import seaborn as sns
```

```python
print(os.listdir("/content/drive/MyDrive/Colab Notebooks/NEU-DET/ images"))
```

```python
 Size = 200
```

```python
train_images =[]
train_labels =[]
for directory_path in glob.glob("/content/drive/MyDrive/Colab Notebooks/NEU-DET/ images/train/images/*"):
  label = directory_path.split("\\")[-1]
  print(label)
  for img_path in glob.glob(os.path.join(directory_path,"*.jpg")):
    print(img_path)
    img= cv2.imread(img_path,cv2.IMREAD_COLOR)
    img = cv2.resize(img,(Size,Size))
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    img = img.reshape(-1)
    train_images.append(img)
    train_labels.append(label)
```

```python
train_images= np.array(train_images)
train_labels= np.array(train_labels)
```

```python
test_images =[]
test_labels =[]
for directory_path in glob.glob("/content/drive/MyDrive/Colab Notebooks/NEU-DET/ images/test/images/*"):
  label = directory_path.split("\\")[-1]
  print(label)
  for img_path in glob.glob(os.path.join(directory_path,"*.jpg")):
    print(img_path)
    img= cv2.imread(img_path,cv2.IMREAD_COLOR)
    img = cv2.resize(img,(Size,Size))
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    img = img.reshape(-1)
    test_images.append(img)
    test_labels.append(label)

test_images= np.array(test_images)
test_labels= np.array(test_labels)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(test_labels)
test_labels_encoded = le.transform(test_labels)
le.fit(train_labels)
train_labels_encoded = le.transform(train_labels)
```

```python
x_train, y_train, x_test, y_test = train_images, train_labels_encoded, test_images, test_labels_encoded
```

```python
from sklearn import svm

svc = svm.SVC(kernel='linear', C=1).fit(x_train, y_train)


```

```python
rbf_svc = svm.SVC(kernel='rbf', gamma=0.5, C=1).fit(x_train, y_train)
```

```python
poly_svc = svm.SVC(kernel='poly', degree=5, C=1).fit(x_train, y_train)
```

```python
y_pred = svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
y_pred = rbf_svc.predict(x_test)
```

```python
 
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
y_pred = poly_svc.predict(x_test)
```
